````python
#Q1. Names of senators whose last name ends with "on"A word - no digits, n number of letters starts with a capital letter ends with ,

command_q1=re.compile(r"[A-Z]\w+(on,)")

for senator in senator_list:
    ans1=re.search(command_q1, senator)
    if ans1!=None:
        print(senator)    
````

````python
#Q2. Get a list of all senators name Thom, Tom, or Thomas
command_q2=re.compile(r",\sT[ho][om]\w?")

for senator in senator_list:
    ans2=re.search(command_q2, senator)
    if ans2!=None:
        print(senator)
````

````python
#Q3. List all "class I" senators
command_q3=re.compile(r"(Class\sI)\W")

for senator in senator_list:
    ans3=re.search(command_q3,senator)
    if ans3!=None:
        print(senator)
````

````python
#Q4. List all class II and III senators

command_q4=re.compile(r"(Class\sII)I?")

for senator in senator_list:
    ans4=re.search(command_q4,senator)
    if ans4!=None:
        print(senator)
````

````python
#Q5. List all independent senators
command_q5=re.compile(r"\(I")

for senator in senator_list:
    ans5=re.search(command_q5,senator)
    if ans5!=None:
        print(senator)
````

````python
#Q6. List all senators whose phone Numbers end with 41
command_q6=(r"-\d+(41)")

for senator in senator_list:
    ans6=re.search(command_q6,senator)
    if ans6!=None:
        print(senator)
````

````python
#Q7. List All senators from a state that begins with A, B or C
command_q7=re.compile(r"[A-C][A-Z]\)")

for senator in senator_list:
    ans7=re.search(command_q7,senator)
    if ans7!=None:
        print(senator)
````

````python
information_list=[]

for senator in senator_list:
    first=re.findall(r",\s[A-Z][a-z]+", senator)
    last=re.findall(r"^[A-Z]\w+", senator)
    initial=re.findall(r"\s[A-Z]\.", senator)
    state=re.findall(r"\s[A-Z][A-Z]\)", senator)
    class_q=re.findall(r"(Class)\s[A-Z]*", senator)
    address=re.findall(r"\d+\s[\w\s\d]+ \(", senator)
    phone=re.findall(r"\(\d+\)\s\d+\-\d+", senator)
    website=re.findall(r"www.+", senator) 
    t=(last,first, initial, state, class_q, address, phone, website)
    information_list.append(t)

information_list
````
